#<center>DECORATORS<center/> 

* Decorators are a very powerful and useful tool in Python since it allows programmers to modify the behaviour of function or class, without actually modifying or editing the original function or method.

* Decorators are of two types: 
       -> User defined Decorators
       -> Built-in Decorators
       






##<center>User defined Decorator</center>

User defined Decorators are of further two types:
        
        * Function Decorators
        * Class Decorators


###<center>Function Decorator</center>

```python

# Normal way to execute Decorator

def update_function(input_function):
  def inner_function():
    y = input_function()
    return y.upper()
  return inner_function

def function():
  return "hello world"
x = function()
print(x, "from function\n")
a = update_function(function)
print(a(), "from update_function\n")


# using @

@update_function
def simple_function():
  return "Muthu pandi Auto"
b = simple_function()
print(b, "from simple_function")
```

```python
# Normal way to  execute Decorator which takes arguments

def check(func):
  def inner(x, y):
    if y == 0:
      return "Enter a valid value"
    return func(x, y)  
  return inner

def div(a, b):
  return a/b
a = check(div)
print(a(4, 2))  
print()

# using @

@check
def d(c, d):
  return c/d
d(5, 0)  
```

```python
# Multiple decorators

def was(func):
  def nus():
    return func() + "yes "
  return nus  
def gus(func):
  def tus():
    return func() + "no "
  return tus
def mum(func):
  def num():
    return func() + "lol "  
  return num

     #  ^
@mum #  |  Direction of execution
@gus #  |
@was #  |

def hum():
  return "hello "
hum()  
```

##<center>Class Decorator<center/>

Class decorators use special method (Call Method) for execution

**Call Method ( $__call__(self)$)**

Call method gives the ability to call the method or function which are defined inside the class, to the objects of that particular class.

If we try to call the method or function of the class by it's object, without mentioning call method, it pop's up an error showing **"Object is not callable"**





```python
# Use of call method in class

class Demo:
  def __init__(self):
    self.height = 160
    self.weight = 50

  def __call__(self):
    print("Function executed with the help of call method")

p = Demo()
print(p.height)
print(p.weight)
p()

```

```python
# without using call method
# pops up error

class Demo:
  def __init__(self):
    self.height = 160
    self.weight = 50

  def demo_func(self):
    print("statement executed from demo_func")

p = Demo()
print(p.height)
print(p.weight)
p()

```

```python
# Class decorator

class Decorator:
  def __init__(self, func):
    self.func = func
  def __call__(self):
    s = self.func()
    return s.upper()

# without using class decoator
def null_func():
  return "la casa de papel"
print(null_func())
print()

# Using class decorator
@Decorator
def null_func():
  return "la casa de papel"
print(null_func())  
```

##<center>@Property</center>
* Property decorator is used to change class methods as attributes.


```python
# when we try to update the value of attribute, it won't reflect on the derivatives of that attribute
# To overcome this we @property method
class Demo:
  def __init__(self, name, age):
    self.name = name
    self.age = age
    self.msg = "hi i am "+self.name+" my age is "+str(age)

n = Demo("davin", 25)
n.name = "bobby"
print(n.name)
n.age = 36
print(n.age)
print(n.msg)
```

```python
class Demo:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def msg(self):
    return "hi i am "+self.name+" my age is "+str(self.age)

n = Demo("davin", 25)
n.name = "bobby"
print(n.name)
n.age = 36
print(n.age)
print(n.msg()) #------> msg is defined as a method, the value of the attribute got updated
```

```python
class Demo:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  @property         #------> property decorator used
  def msg(self):
    return  "hi i am "+self.name+" my age is "+str(self.age)

n = Demo("davin", 25)
n.name = "bobby"
print(n.name)
n.age = 36
print(n.age)
print(n.msg) #------> msg is represented as an attribute
```

```python
# Using setter from property method
class Demo:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  @property        
  def msg(self):
    return  "hi i am "+self.name+" my age is "+str(self.age)
  @msg.setter
  def msg(self,msg):
    x = msg.split(" ")
    self.name = x[0]
    self.age = x[-1]
n = Demo("davin", 25)
n.name = "bobby"
print(n.name)
n.age = 36
print(n.age)
n.msg = "delson is 15"
print(n.msg)
```
